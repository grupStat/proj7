### Zadanie 2: Estymacja funkcją jądrową [16 pkt]

#### a) Estymacji funkcjami jądrowymi od podstaw

Przygotuj funkcję `simple_kde_estimation()`, która będzie dokonywała estymacji nieparametrycznej metodą funkcji jądrowych z wykorzystaniem jądra w postaci rozkładu normalnego o odchyleniu normalnym $\sigma$ zgodnie z informacjami zawartymi w docstring funkcji oraz schematem z załączonego do notatnika zestawu wzorów.

Poza zwracaniem danych określonych w docstring funkcji powinna ona generować wykres zawierający wyestymowaną funkcję (wykres liniowy z wypełnionym polem pod krzywą) oraz dodaną wizualną reprezentacją zestawu danych w postaci zbioru symboli (| lub +) zlokalizowanych poniżej wykresu.

Uwagi do wykonania zadania:
 - Punkty estymacji wygeneruj za pomocą funkcji `numpy.linspace()` lub `numpy.arange()`.
 - Wartości gęstości prawdopodobieństwa oblicz za pomocą odpowiedniej funkcji z modułu `scipy.stats.norm`.
 - Przygotowując figurę kieruj się wytycznymi z początku notatnika.

```python
def simple_kde_estimation(data, estimation_points=1000, sigma=1.0):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierająca domyślny indeks i pojedynczą kolumnę "X" z wartościami próby losowej.
    estimation_points: int
        Liczba punktów estymacji funkcji gęstości prawdopodobieństwa.
    sigma: float
        Odchylenie standardowe rozkładu normalnego wykorzystywanego jako jądro funkcji.

    Returns
    -------
    pdf: DataFrame
        Tablica zawierająca następujące kolumny:
        X - uszeregowane rosnąco punkty estymacji funkcji gęstości prawdopodobieństwa,
        KDE - wyestymowane wartości funkcji gęstości prawdopodobieństwa.
    """
    sample = data['X'].values

    # Tworzenie siatki punktów do estymacji
    x_min = sample.min() - 3 * sigma
    x_max = sample.max() + 3 * sigma
    x_grid = np.linspace(x_min, x_max, estimation_points)


    # Obliczanie estymacji KDE
    kde_values = np.zeros_like(x_grid)
    for xi in sample:
        kde_values += sp.stats.norm.pdf(x_grid, loc=xi, scale=sigma)
    kde_values /= len(sample)

    # Rysowanie wykresu
    plt.figure(figsize=(10, 6))
    plt.plot(x_grid, kde_values, color='blue', lw=2, label='KDE')
    plt.fill_between(x_grid, kde_values, color='blue', alpha=0.3)
    plt.title('Estymacja jądrowa gęstości prawdopodobieństwa')
    plt.xlabel('Wartość')
    plt.ylabel('Gęstość')

    # Dodanie znaczników danych poniżej osi X
    y_min = -0.01 * kde_values.max()
    plt.plot(sample, np.full_like(sample, y_min), '|', color='black', markersize=12)

    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Przygotowanie wyników jako DataFrame
    pdf = pd.DataFrame({'X': x_grid, 'KDE': kde_values})
    return pdf
```

```python
# Komórka testowa
test_results_1 = simple_kde_estimation(sample, estimation_points=100, sigma=0.2)
assert type(test_results_1) == pd.DataFrame
assert test_results_1.shape == (100, 2)
assert list(test_results_1.columns) == ['X', 'KDE']
assert np.all(np.isclose(test_results_1.min().values, [2.94233698e+00, 3.32813073e-04]))
assert np.all(np.isclose(test_results_1.max().values, [6.98176057, 0.54688772]))
assert np.all(np.isclose(test_results_1.mean().values, [4.96204877, 0.24508447]))

test_results_2 = simple_kde_estimation(sample, estimation_points=250, sigma=0.35)
assert type(test_results_2) == pd.DataFrame
assert test_results_2.shape == (250, 2)
assert list(test_results_2.columns) == ['X', 'KDE']
assert np.all(np.isclose(test_results_2.min().values, [2.49233698e+00, 2.69768435e-04]))
assert np.all(np.isclose(test_results_2.max().values, [7.43176057, 0.47782523]))
assert np.all(np.isclose(test_results_2.mean().values, [4.96204877, 0.20164296]))
```

#### b) Dedykowane funkcje

Korzystając z klasy `KernelDensity` wchodzącej w skład pakietu scikit-learn dokonaj estymymacji rozkładu zestawu danych zawartego w zmiennej `sample`.

Wyniki estymacji przedstaw w postaci wykresów zawierających wyestymowaną funkcję (wykres liniowy z wypełnionym polem pod krzywą) oraz dodaną wizualną reprezentacją zestawu danych w postaci pionowych kresek zlokalizowanych poniżej wykresu.

Wykresy zestaw na dwóch figurach (2x3 pola na wykres). Przygotowując figury kieruj się wytycznymi z początku notatnika. Podczas wykonywania wykresów:
 - Figura 1:
     - na każdym z wykresów przedstaw wynik estymacji z wykorzystaniem innego jądra funkcji,
     - ustaw wartość parametru odpowiadającego za szerokość jądra funkcji na 0.1,
     - ustaw identyczne zakresy osi dla wszystkich wykresów.
 - Figura 2:
     - na każdym z wykresów przedstaw wynik estymacji z wykorzystaniem innego jądra funkcji,
     - ustaw wartość parametru odpowiadającego za szerokość jądra funkcji tak, aby wynik pracy funkcji dobrze oddawał charakter rozkładu populacji z której została pobrana próba losowa,
     - ustaw identyczne zakresy osi dla wszystkich wykresów. 

```python
x_min, x_max = sample['X'].min() - 1, sample['X'].max() + 1
x_grid = pd.DataFrame(np.linspace(x_min, x_max, 1000), columns=["X"])

# Kernels to compare
kernels = ['gaussian', 'tophat', 'epanechnikov', 'exponential', 'linear', 'cosine']

# Figure 1 - fixed bandwidth (0.1)
plt.figure(figsize=(18, 10))
for i, kernel in enumerate(kernels, 1):
    plt.subplot(2, 3, i)
    
    # KDE estimation
    kde = KernelDensity(kernel=kernel, bandwidth=0.1).fit(sample[['X']])
    log_dens = kde.score_samples(x_grid)
    dens = np.exp(log_dens)
    
    # Plot KDE
    plt.plot(x_grid, dens, color='blue', lw=2)
    plt.fill_between(x_grid.values.flatten(), dens, color='blue', alpha=0.2)
    
    # Data ticks
    plt.eventplot(sample['X'], colors='red', linelengths=0.1, lineoffsets=-0.02)
    
    plt.title(f'Kernel: {kernel}\nBandwidth=0.1')
    plt.xlim(x_min, x_max)
    plt.ylim(0, 1)
    plt.grid(True)

plt.tight_layout()
plt.suptitle('Density estimation with different kernels (bandwidth=0.1)', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(18, 10))
for i, kernel in enumerate(kernels, 1):
    plt.subplot(2, 3, i)
    
    # Set optimal bandwidths for each kernel
    optimal_bw = {
        'gaussian': 0.25,
        'tophat': 0.3,
        'epanechnikov': 0.25,
        'exponential': 0.2,
        'linear': 0.35,
        'cosine': 0.25
    }[kernel]
    
    kde = KernelDensity(kernel=kernel, bandwidth=optimal_bw).fit(sample[['X']])
    log_dens = kde.score_samples(x_grid)
    dens = np.exp(log_dens)
    
    # Plot KDE
    plt.plot(x_grid, dens, color='green', lw=2)
    plt.fill_between(x_grid.values.flatten(), dens, color='green', alpha=0.2)
    
    # Data ticks
    plt.eventplot(sample['X'], colors='black', linelengths=0.1, lineoffsets=-0.02)
    
    plt.title(f'Kernel: {kernel}\nBandwidth={optimal_bw:.2f}')
    plt.xlim(x_min, x_max)
    plt.ylim(0, 1)
    plt.grid(True)

plt.tight_layout()
plt.suptitle('Estymacja gęstości z różnymi jądrami', y=1.02)
plt.show()
```

Przygotowane w poprzednich zadaniach wykresy są wizualną reprezentacją tego samego zestawu danych, dobór metody oraz jej parametrów sparawia jednak, że nie wyglądają one identycznie.

Plusy i minusy metod estymacji

Histogramy

Plusy:

Proste w implementacji i interpretacji
Szybkie obliczeniowo
Intuicyjne dla osób niebędących specjalistami w statystyce
Dobrze pokazują ogólny kształt rozkładu

Minusy:

Wrażliwe na dobór liczby przedziałów (bins)
Mogą ukrywać szczegóły rozkładu przy złym doborze parametrów
Tworzą sztuczne "schodki" w miejscach gdzie rozkład jest ciągły
Nie zapewniają płynnego przejścia między przedziałami

Estymacja jądrowa (KDE)

Plusy:

Daje płynną estymację gęstości
Lepiej oddaje charakter ciągłych rozkładów
Mniej wrażliwa na wybór punktu startowego (w porównaniu do histogramu)
Pozwala na użycie różnych jąder, które mogą lepiej pasować do danych

Minusy:

Bardziej złożona obliczeniowo
Wymaga doboru odpowiedniego jądra i parametru wygładzania (bandwidth)
Może być trudniejsza w interpretacji dla osób nieznających metody
Przy zbyt małym wygładzaniu może pokazywać artefakty (przeuczenie)
Wpływ parametrów na wyniki

Liczba przedziałów w histogramie:

Zbyt mała liczba - tracimy szczegóły rozkładu, może ukryć multimodalność
Zbyt duża liczba - pojawia się nadmierna zmienność, trudno dostrzec ogólny kształt
Szerokość przedziałów w histogramie:
Zbyt szerokie - rozkład wydaje się zbyt wygładzony
Zbyt wąskie - wykres staje się "postrzępiony"
Położenie granic przedziałów:
Może znacząco wpłynąć na wygląd histogramu, szczególnie przy małej liczbie punktów
Nierozważny wybór może sztucznie ukryć lub uwypuklić pewne cechy danych
Wybór jądra w KDE:
Różne jądra (gaussowskie, epanechnikow, trójkątne itp.) mają różne właściwości wygładzania
Jądro gaussowskie jest najbardziej uniwersalne, ale inne mogą lepiej pasować do specyficznych rozkładów
Szerokość jądra (bandwidth) w KDE:
Zbyt mała - estymator jest zbyt "rozdrobniony", pokazuje lokalne fluktuacje zamiast ogólnego kształtu
Zbyt duża - tracimy szczegóły rozkładu, estymator jest zbyt wygładzony
Optymalna wartość zależy od rozkładu danych i liczby obserwacji

Podsumowanie:

Dobór parametrów ma kluczowe znaczenie dla jakości estymacji. W przypadku histogramu najważniejszy jest wybór liczby i szerokości przedziałów, podczas gdy w KDE - wybór jądra i parametru wygładzania. Właściwy dobór tych parametrów pozwala na uzyskanie estymatora, który dobrze oddaje prawdziwy rozkład danych, nie będąc jednocześnie ani zbyt wygładzonym (co prowadzi do utraty informacji), ani zbyt "rozdrobnionym" (co pokazuje głównie fluktuacje losowe).


### Zadanie 4: Estymacja na podstawie dystybuanty [8 pkt]

Przygotuj funkcję `cdf_estimation()`, która będzie dokonywała estymacji dystrybuanty rozkładu w oparciu o dystrybuantę empiryczną oraz statystykę wyznaczaną z rozkładu Kołmogorowa-Smirnowa zgodnie z informacjami zawartymi w docstring funkcji oraz schematem z załączonego do notatnika zestawu wzorów.

Poza zwracaniem danych określonych w docstring funkcji powinna ona generować wykres liniowy dystrybuanty empirycznej z zaznaczonym przedziałem ufności oraz tytułem figury `Przedział ufności dystrybuanty rozkładu na poziomie istotności 1-alpha=A`, gdzie wartość A jest uzupełnianie na podstawie wartości argumentu `alpha`.

Uwagi do wykonania zadania:
 - Wartości statystyki testowej oblicz za pomocą odpowiedniej funkcji z modułu `scipy.stats.kstwo`.

```python
from scipy.stats import kstwo

def cdf_estimation(data, alpha=0.5):
    
    x = np.sort(data["X"].values)
    n = len(x)
    ecdf = np.arange(1, n + 1) / n

    d_alpha = kstwo.ppf(1 - alpha, n) / np.sqrt(n)
    lower = np.maximum(ecdf - d_alpha, 0)
    upper = np.minimum(ecdf + d_alpha, 1)

    # Wykres
    plt.figure(figsize=(8, 5))
    plt.step(x, ecdf, where='post', label="Dystrybuanta empiryczna", color='blue')
    plt.step(x, lower, where='post', linestyle='--', label="Dolna granica ufności", color='red')
    plt.step(x, upper, where='post', linestyle='--', label="Górna granica ufności", color='green')

    plt.title(f"Przedział ufności dystrybuanty rozkładu na poziomie istotności 1-alpha={1 - alpha:.2f}")
    plt.xlabel("Wartości próbki")
    plt.ylabel("Dystrybuanta empiryczna")
    plt.ylim(-0.05, 1.05)
    plt.legend()
    plt.grid(True)
    plt.show()

    return pd.DataFrame({
        "X": x,
        "F_d": ecdf,
        "F_d_min": lower,
        "F_d_max": upper
    })
```

```python
# Komórka testowa
test_results_3 = cdf_estimation(sample, 0.05)
assert type(test_results_3) == pd.DataFrame
assert test_results_3.shape == (130, 4)
assert list(test_results_3.columns) == ['X', 'F_d', 'F_d_min', 'F_d_max']
assert np.all(np.isclose(test_results_3.min().values, [3.54233698, 0.00769231, 0., 0.12544548]))
assert np.all(np.isclose(test_results_3.max().values, [6.38176057, 1., 0.88224683, 1.]))
assert np.all(np.isclose(test_results_3.mean().values, [4.9551826 , 0.50384615, 0.3925793 , 0.61420722]))

test_results_4 = cdf_estimation(sample, 0.02)
assert type(test_results_4) == pd.DataFrame
assert test_results_4.shape == (130, 4)
assert list(test_results_4.columns) == ['X', 'F_d', 'F_d_min', 'F_d_max']
assert np.all(np.isclose(test_results_4.min().values, [3.54233698, 0.00769231, 0., 0.13935689]))
assert np.all(np.isclose(test_results_4.max().values, [6.38176057, 1., 0.86833542, 1.]))
assert np.all(np.isclose(test_results_4.mean().values, [4.9551826 , 0.50384615, 0.380346, 0.62633351]))
```
